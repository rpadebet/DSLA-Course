```r
Let us normalize the variables and see if we have the same kind of difference 
in including and removing the intercept term
```

```r
mtcars$mpg_norm <- (mtcars$mpg - mean(mtcars$mpg))/sd(mtcars$mpg)
mtcars$wt_norm <- (mtcars$wt - mean(mtcars$wt))/sd(mtcars$wt)

mdl3<-lm(mpg_norm ~ wt_norm,data = mtcars)
summary(mdl3)
```

```r
plot(wt_norm, mpg_norm, main="Scatterplot Example",xlab="Car Weight ", ylab="Miles Per Gallon ", pch=19)
# Add fit line
abline(mdl3, col="red") # regression line (y~x) 
```

```r
That looks good when we included the intercept.

What happens if we remove the intercept as before?
```

```r

mdl4<-lm(mpg_norm ~ wt_norm -1,data = mtcars)
summary(mdl4)
```

```r
plot(wt_norm, mpg_norm, main="Scatterplot Example",xlab="Car Weight ", ylab="Miles Per Gallon ", pch=19)
# Add fit line
abline(mdl4, col="red") # regression line (y~x) 
```

```r
We still get the same model once we normalize the data even if we remove the intercept
```
